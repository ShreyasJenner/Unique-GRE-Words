### Read vince's file which is in csv format and then create a list to store data column wise

```python
import pandas
df1 = pandas.read_csv("vince.csv")                                    #read vince's file which is in csv form 
pandas.set_option('display.max_rows',1400)
pandas.set_option('display.max_columns',20)
df2 = pandas.read_csv("vince.csv",header=None)                        
df2 = df2.drop(df2.index[0:1],0)
df3 = df2.loc[1:1263,1:8] 
vince=[]                                                             #list to store only words from vince                                                                                 
for i in df3.loc[:,1]:
    vince.append(i)

greg=[]                                                              #list to store only words from gregmat
for j in df3.loc[:,2]:
    greg.append(j)

magoosh=[]                                                          #lists to store only words from magoosh
for a in df3.loc[:,3]:
    magoosh.append(a)

manhattan=[]                                                        #lists to store only words from manhanttan
for b in df3.loc[:,4]:
    manhattan.append(b)

prep=[]
for c in df3.loc[:,5]:
    prep.append(c)
    
powerscore=[]
for d in df3.loc[:,6]:
    powerscore.append(d)

gbasic=[]
for e in df3.loc[:,7]:
    gbasic.append(e)

gadvance=[]
for f in df3.loc[:,8]:
    gadvance.append(f)
```

### the below code removes all null boxes from the column so that no errors occur while reading the lists

```python
greg = [x for x in greg if pandas.isnull(x)==False]                     
magoosh = [x for x in magoosh if pandas.isnull(x)==False]
manhattan = [x for x in manhattan if pandas.isnull(x)==False]
prep =  [x for x in prep if pandas.isnull(x)==False]
powerscore = [x for x in powerscore if pandas.isnull(x)==False]
gbasic = [x for x in gbasic if pandas.isnull(x)==False]
gadvance = [x for x in gadvance if pandas.isnull(x)==False]
```

### create a list to store all edited lists and then create each lists conjugate list

```python
all = [greg,magoosh,manhattan,prep,powerscore,gbasic,gadvance]

wvince = list((greg+magoosh+manhattan+prep+powerscore+gbasic+gadvance))             #wvince has words from all lists except vince
wgreg = list((vince+magoosh+manhattan+prep+prep+powerscore+gbasic+gadvance))        
wmagoosh = list((vince+greg+manhattan+prep+powerscore+gbasic+gadvance))
wmanhattan = list((vince+greg+magoosh+prep+powerscore+gbasic+gadvance))
wprep = list(vince+greg+magoosh+manhattan+powerscore+gbasic+gadvance)
wpowerscore = list((vince+greg+magoosh+manhattan+prep+gbasic+gadvance))
wgbasic = list((vince+greg+magoosh+manhattan+prep+powerscore+gadvance))
wgadvance = list((vince+greg+magoosh+manhattan+prep+powerscore+gbasic))

wall = [wgreg,wmagoosh,wmanhattan,wprep,wpowerscore,wgbasic,wgadvance]
```

### function that stores only unique words 

```python
def inter(a,b):
    temp =[x for x in a if x not in b]    # Function of code that allows to print only uncommon elements
    return temp
```

### iterates lists with their respective conjugate lists and then adds only those words that are unique in the original list

```python
b=[]
for i,j in zip(all,wall):
    b.append(inter(i,j))              #for loop that iterates lists with each other
print(b)
```

### stores list as a dataframe 

```python
df1 = pandas.DataFrame(b)                       #convert list to a dataframe and store it
df1=df1.T                                       #transposes index and column of data frames
df1
```
